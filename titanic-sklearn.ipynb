```python
import numpy as np 
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/kaggle/input/titaniccsv/titanic.csv')
df.head()
```

```python
df = df.dropna(subset = ['Sex', 'Age', 'Pclass', 'Fare', 'Parch'])
df['Sex'] = df.Sex.replace(['male', 'female'], [0, 1])
df.shape
```

```python
fare = df.Fare.values.reshape(-1, 1)
fare_scaled = MinMaxScaler().fit_transform(fare)
df['Fare'] = pd.DataFrame(fare_scaled)
df.Fare.shape
```

```python
df.info()
```

```python
df['Fare'] = df['Fare'].fillna(value = 0)
```

```python
X_train = df[['Sex', 'Age', 'Pclass', 'Fare', 'Parch']].astype('float64')
X_train.shape
```

```python
X_train.info()
```

```python
y_train = df['Survived']
y_train.describe()
```

```python
df_test = pd.read_csv('/kaggle/input/titaniccsv/test.csv')
df_test.head()
```

```python
df_test['Sex'] = df.Sex.replace(['male', 'female'], [0, 1])
df_test.head()
```

```python
fare = df_test.Fare.values.reshape(-1, 1)
fare_scaled = MinMaxScaler().fit_transform(fare)
df_test['Fare'] = pd.DataFrame(fare_scaled)
df_test.Fare.describe()
```

```python
X_test = df_test[['Sex', 'Age', 'Pclass', 'Fare', 'Parch']].astype('float64')
X_test = X_test.fillna(method = 'pad')
X_test.shape
```

```python
X_test.info()
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
pred = pd.DataFrame(pred, columns = ['Survived'])
sub = pd.concat([df_test.PassengerId, pred], axis=1)
sub.to_csv('submission.csv', index=False)
```
